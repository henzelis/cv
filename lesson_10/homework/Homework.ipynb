### Step 1
Decide what video you are going to use for this homework, select an object and generate the template. You can use any video you want (your own, from Youtube, etc.)
and track any object you want (e.g. a car, a pedestrian, etc.).


```python
import cv2
from matplotlib import pyplot as plt

file = 'E:\Python\cv\lesson_10\hightway.mp4'
    
# Load video file
video = cv2.VideoCapture(file)
ok, frame = video.read()

# Setting the coordinates of monitoring object
x1 = 642
x2 = 707
y1 = 90
y2 = 118

# Getting box parameters
width = x2 - x1
height = y2 - y1

# Crating bbox
bbox = (x1, y1, width, height)

```

### Step 2
Initialize a tracker (e.g. KCF).

```python
#Initialization tracker
# Creating a Tracker list with types
tracker_types = ['MIL','KCF', 'CSRT']
tracker_type = tracker_types[1]

if tracker_type == 'MIL':
    tracker = cv2.TrackerMIL_create()

if tracker_type == 'KCF':
    tracker = cv2.TrackerKCF_create()

if tracker_type == "CSRT":
    tracker = cv2.TrackerCSRT_create()
    
ok = tracker.init(frame, bbox)
```

### Step 3
Run the tracker on the video and the selected object. Run the tracker for around 10-15 frames.

### Step 4
For each frame, print the bounding box on the image and save it.

```python
for frames in range(1, 10):
    ret, frame = video.read()
    ok, bbox = tracker.update(frame)
    print(ok, bbox)

    x1, y1 = bbox[0], bbox[1]
    width, height = bbox[2], bbox[3]
    cv2.rectangle(frame, (int(x1), int(y1)), (int(x1) + int(width), int(y1) + int(height)), (0, 255, 0), 2)
    cv2.putText(frame, tracker_type + " Tracker", (100, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.75, (50, 170, 50), 2);
    plt.imshow(frame)
    plt.show(), plt.draw()
    plt.waitforbuttonpress(0.1)
    plt.clf()
```

### Step 5
Select a different tracker (e.g. CSRT) and repeat steps 2, 3 and 4.

```python
#Initialization tracker
tracker_type = tracker_types[2]
ok = tracker.init(frame, bbox)


for frames in range(1, 10):
    ret, frame = video.read()
    ok, bbox = tracker.update(frame)
    print(ok, bbox)

    x1, y1 = bbox[0], bbox[1]
    width, height = bbox[2], bbox[3]
    cv2.rectangle(frame, (int(x1), int(y1)), (int(x1) + int(width), int(y1) + int(height)), (0, 255, 0), 2)
    cv2.putText(frame, tracker_type + " Tracker", (100, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.75, (50, 170, 50), 2);
    plt.imshow(frame)
    plt.show(), plt.draw()
    plt.waitforbuttonpress(0.1)
    plt.clf()
```
